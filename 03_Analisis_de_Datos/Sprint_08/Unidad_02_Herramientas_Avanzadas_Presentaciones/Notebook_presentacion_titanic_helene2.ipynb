```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic = pd.read_csv("./data/titanic.csv")
```

```python
#1. Mortalidad alta en el accidente del Titanic. 

data = titanic["alive"].value_counts()
fig,ax = plt.subplots(1,1,figsize = (4,4))
ax.pie(data.values,
        labels= ("Fallecidos","Sobreviviente"),
        colors = ("darkgrey", "lightgreen"),
        autopct='%.f%%', startangle= 0);
ax.set_title("Mortalidad")
fig.savefig("./data/1mortalidad.png")
```

```python
#2 mujeres y niños primeros - mejorar aun el % proportion, index
child = titanic.loc[titanic["who"]=="child"]
women = titanic.loc[titanic["who"]=="woman"]
men = titanic.loc[titanic["who"]=="man"]
color = ("darkgrey", "beige")

fig,axs = plt.subplots(nrows=1,ncols=3, figsize=(15,4), sharey= True)
fig.suptitle("Mujeres y niños, primeros evacuados");

valores = child.alive.value_counts(normalize = True) * 100
sns.barplot(x = valores.sort_index(), y = valores.values, hue = valores.index,palette= ("lightgreen","grey"), ax = axs[0])
axs[0].set_title("% Niños")



valores = women.alive.value_counts(normalize = True) * 100
sns.barplot(x = valores.sort_index(), y = valores.values, hue = valores.index, palette= ("lightgreen","grey"),ax = axs[1])
axs[1].set_title("% mujeres")


valores = men.alive.value_counts(normalize = True) * 100
sns.barplot(x = valores.sort_index(), y = valores.values, hue = valores.index, palette= ("lightgreen","grey"), ax = axs[2])
axs[2].set_title("% hombres");

fig.savefig("./data/2mujeres_primero.png")
```

```python
#3 clases - ok podria juntar imagenes


data = titanic["class"].value_counts()
fig,ax = plt.subplots(1,1,figsize = (4,4))
ax.pie(data.values, labels= ("Tercera","Primera", "Segunda") ,   
        colors = ("orchid","skyblue", "yellow")    ,
        autopct='%.0f%%', startangle= 90);
ax.set_title("Reparto de classes total")
my_circle = plt.Circle((0,0),0.4, color= "white")
ax.add_artist(my_circle)
fig.savefig("./data/3clase_total.png")


fallecidos= titanic.loc[titanic["alive"]=="no"]
data = fallecidos["class"].value_counts()
fig,ax = plt.subplots(1,1,figsize = (4,4))
ax.pie(data.values,        labels= ("Tercera","Primera", "Segunda") ,  
         colors = ("orchid","skyblue", "yellow")    , 
         autopct='%.0f%%', startangle= 90);
ax.set_title("Reparto de classes de fallecidos")
my_circle = plt.Circle((0,0),0.4, color= "white")
ax.add_artist(my_circle)
fig.savefig("./data/3clase_fallecidos.png")
```

```python
titanic_woman = titanic.loc[titanic["who"]=="woman"]
titanic_child = titanic.loc[titanic["who"]=="child"]


fig, axs= plt.subplots(nrows=1, ncols=1, figsize = (10,4))
fig.suptitle(" % Mujeres sobrevivientes",size=20)

frecuencias_class= titanic_woman["alive"].value_counts()
df_frecuencias_class = titanic_woman["alive"].value_counts().reset_index() 
df_frecuencias_class.columns= ["categorias","frecuencias_class"]
sns.scatterplot(x="categorias", y= [1]*len(frecuencias_class), markers= False, hue="categorias", palette = ("lightgreen","grey"), data = df_frecuencias_class, size= "frecuencias_class", legend= False, sizes= (100,2000))
axs.set_xlabel("")
for clase, valor in frecuencias_class.items():
    axs.text(clase,1.03, f"{valor}({round(valor*100/frecuencias_class.sum())}%)")
axs.set_facecolor("None")
axs.yaxis.set_ticks([])
fig.savefig("./data/4mujeres.png")
```
